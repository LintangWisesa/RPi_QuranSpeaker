# Surah Al-Qur'an

Daftar surah dalam Al-Qur'an, di-scraping dari [Wikipedia Indonesia](https://id.wikipedia.org/wiki/Surah).

```python
import numpy as np
import pandas as pd
```

```python
url = 'https://id.wikipedia.org/wiki/Surah'
df = pd.read_html(url)
```

```python
dfSurah = df[1].copy()
dfSurah
```

```python
for i in range(len(dfSurah)):
    print(f'<option value="{i+1}">{dfSurah.iloc[i]["No."]}. {dfSurah.iloc[i]["Nama Surah"]} {dfSurah.iloc[i]["Bahasa Arab"]}</option>')
```
